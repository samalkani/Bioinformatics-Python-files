# Manipulating DNA Strings

### Finding the longest common prefix between two strings using a user defined function

```python
def longestCommonPrefix(s1, s2):
    i = 0
    while i < len(s1) and i < len(s2) and s1[i] == s2[i]:
        i += 1
    return s1[:i]

longestCommonPrefix('ACCATGT','ACCAGAC')
    
```

### Writing a function to see whether two strings match exactly

```python
def match(s1, s2):
    
    if not len(s1) == len(s2):
        return False
    
    for i in range(len(s1)):
        if not s1[i] == s2[i]:
            return False
    
    return True

match('ATGCT','ATGCT')
```

### Using match() function for two strings that are not the same

```python
match('ATCGT','AACCG')
```

```python
match('ATCGT','ATCGTG')
```

### Using a dictionary to give the reverse complement of a DNA sequence

```python
complement = {'A':'T', 'C':'G', 'G':'C', 'T':'A'}
```

### Key 'A' in the dictionary outputs 'T'

```python
complement['A']
```

### Key 'C' in the dictionary outputs 'G'

```python
complement['C']
```

### Using the above dictionary in our user defined function

#### Notice t = complement[base] + t gives the reverse complement instead of
####  doing t = t + complement[base] which does not!

```python
def reverseComplement(s):
    complement = {'A':'T', 'C':'G', 'G':'C', 'T':'A'}
    t = ''
    for base in s:
        t = complement[base] + t
    return t

reverseComplement('ACCGTCG')
    
```
