### Create a function to read a genome from a fasta file

#### (Note: filename=r'filepath' to file since you are passing in a string)

```python
def readGenome(filename):
    genome = ''
    with open(filename, 'r') as f:
        for line in f:
            if not line[0] == '>':
                genome += line.rstrip()
                
    return genome


genome = readGenome(r'lambda_virus.fa')
genome[:100]
```

### Confirm that the entire file has been read using len() function

```python
len(genome)
```

### Counting the frequency of each base using a dictionary

```python
counts = {'G':0, 'T':0, 'A':0, 'C':0}
for base in genome:
    counts[base] += 1
print(counts)
```

### Counting bases in a DNA sequence as above using a Python module (Counter Object produced)

```python
import collections
collections.Counter(genome)
```
